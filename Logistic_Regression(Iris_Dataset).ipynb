## **Logistic_Regression Model/Classification Model** 

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
x=iris.drop(['target','target_label'],axis=1)
x.head()
```

```python
y = iris['target']
y.head()
```

## **Train Test Split**

You always want to evaluate your final model on a test set that hasn't been used at all in the training process. So, we'll split off a test set here.

_(Note : This might be a little less true if you use cross validation,but is still considered as a best practice.)_

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=.3, random_state=5)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred= model.predict(x_test)
y_pred
```

```python
(y_pred == y_test).values
```

```python
np.mean(y_pred == y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy=confusion_matrix(y_test,y_pred)
```

```python
accuracy
```

```python
from sklearn.metrics import accuracy_score, precision_score
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(f'Accuracy of our model is : {round(accuracy*100,2)} %')
```

```python
precision=precision_score(y_test,y_pred, average='macro')
precision
```

```python
from sklearn.metrics import classification_report
```

```python
print(f'Here is our classification report : \n {classification_report(y_test,y_pred)}')
```
